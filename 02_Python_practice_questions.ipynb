#### 7.Even or Odd Checker
##### Ask for a number.Tell if it's even or odd.Otherwise Exit with exit.

```python
while True:
    user_input=input("Enter number(or exit):")
    if user_input=="exit":
        print("Done!")
        break
    if user_input.isdigit():
        num=int(user_input)
        if num%2==0:
            print("Even Number")
        else:
            print("Odd Number")
    else:
        print("Enter a Valid Number!")
```

### Dictionary Comprehension:

#### 11.Given a list of product prices: `product_prices = [("laptop", 1200), ("keyboard", 75), ("mouse", 25)]`. Use dictionary comprehension to create a dictionary where product names are keys and prices are values.

```python
product_prices = [("laptop", 1200), ("keyboard", 75), ("mouse", 25)]
products={product_names:prices for product_names,prices in product_prices}
print(products)
```

#### 12.Given a dictionary of item quantities: `items_quantities = {"pen": 10, "notebook": 5, "eraser": 0, "pencil": 12}`. Use dictionary comprehension to create a new dictionary `in_stock_items` containing only items with a quantity greater than 0.

```python
items_quantities = {"pen": 10, "notebook": 5, "eraser": 0, "pencil": 12}
in_stock_items={product_names:quantity for product_names,quantity in items_quantities.items() if quantity>0}
print(in_stock_items)
```
